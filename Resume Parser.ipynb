# Resume Parser

```python
import docx2txt
import nltk
from nltk.tokenize import regexp_tokenize
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('averaged_perceptron_tagger')
stop = set(stopwords.words('english'))
wnl = WordNetLemmatizer()
pattern = r'(\w+)'
```

```python
def func_lemmatize(words):
    lemmatized = []
    for word, tag in pos_tag(words):
        wntag = tag[0].lower()
        if wntag in ['n', 'v']:
            lemma = wnl.lemmatize(word, wntag)
            lemmatized.append((lemma, wntag))
    return lemmatized

def func_cleanup(doc):
    string_doc = docx2txt.process(doc)
    tokenized_doc = regexp_tokenize(string_doc, pattern)
   
    tokenized_doc2 = []
    for word in tokenized_doc:
        word  = word.lower()
        if word not in stop:
            tokenized_doc2.append(word)
            
    clean_list = func_lemmatize(tokenized_doc2)
    clean_dict = {'n': [] , 'v': []}
    for pair in clean_list:
        if pair[1] == 'n':
            clean_dict['n'].append(pair[0]) 
        else: 
            clean_dict['v'].append(pair[0])
            
    for key in clean_dict:
        clean_dict[key] = list(set(clean_dict[key]))
    
    
    return clean_dict

                            
```

```python
jdesc = func_cleanup('jd-analyst.docx')
jdesc
```

```python
res = func_cleanup('res-analyst.docx')
```

```python
power_verbs = docx2txt.process('power-verbs.docx')
```

```python
powerverbs_clean = regexp_tokenize(power_verbs, pattern)
powerverbs_final = [word.lower() for word in powerverbs_clean]
```

```python
powerverbs_final
```

```python
jdesc_nouns = jdesc['n']
res_nouns = res['n']
```

```python
counter = 0

for word in jdesc_nouns:
    if word in res_nouns:
        counter += 1

nscore = counter / len(res_nouns)
print('Your noun match score is', nscore)
```

```python
jdesc_verbs = jdesc['v']
res_verbs = res['v']

```

```python
counter = 0

for word in res_verbs:
    if word in jdesc_verbs:
        counter+=1

vscore = counter / len(jdesc_verbs)

print('Your verb match score is', vscore)
```

```python
counter = 0

for word in res_verbs:
    if word in powerverbs_final:
        counter += 1

power_score = counter / 100
```

```python
totalscore = (nscore + vscore)/2 + power_score
print('Your overall score is', totalscore)
```

```python
scores_dict = {(0.50,1.01):['A','Well done'],(0.30,0.50):['B','Not bad'],(0,0.30):['C','Not good enough bro']}
```

```python
for key in scores_dict:
    if totalscore >= key[0] and totalscore < key[1]:
        grade = scores_dict[key][0]
        comment = scores_dict[key][1]
        
print(f'Your grade is {grade}. {comment}!')
```
